# 1 · Quickstart

**roadstyle** turns a GeoDataFrame of roads into a styled, interactive map — an OSM-style *geometry sandwich* (a coloured fill over a wider casing) so junctions read cleanly.

You need just two things: a **road-class column** (default `highway`) and **line geometry**.

```python
import pathlib
import geopandas as gpd
import roadstyle as rs

# A GeoDataFrame of road edges. roadstyle only needs a *road-class* column (default "highway")
# and line geometry in any CRS — it reprojects to web coordinates (EPSG:4326) for you.
edges = gpd.read_file(pathlib.Path("data") / "sundbyberg_edges.gpkg")
print(f"{len(edges):,} edges, CRS {edges.crs}")
edges[[c for c in edges.columns if c != edges.geometry.name]].head(3)
```

## Render with one call
`render_edges` returns a folium map. In Jupyter, the last expression displays it inline.

```python
m = rs.render_edges(edges, theme="dark")   # high-saturation palette on a dark base map
m
```

## Themes & palettes
Two palettes — `highsat` (high-contrast) and `carto` (classic OSM) — and three themes: `light`, `dark`, `satellite`.

```python
rs.render_edges(edges, palette="carto", theme="light")
```

## Save to a standalone HTML file
`.save()` writes a self-contained page you can open in any browser or drop into a site.

```python
# Maps/files in this guide are written to ./output/ (git-ignored), so nothing clutters the repo.
out = pathlib.Path("output"); out.mkdir(exist_ok=True)
rs.render_edges(edges, theme="dark").save(out / "quickstart.html")
print("wrote", out / "quickstart.html")
```

That's the whole loop: **load → `render_edges` → `.save`**. The next notebooks add data-driven colour, filtering, and web embedding.
